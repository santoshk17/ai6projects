```python
import os
import keras
import pandas as pd
import shutil
```

```python
rootPath="D:\\data\\all\\"
```

```python
os.chdir(rootPath)
data = pd.read_csv("D:\\data\\all\\labels.csv") 
os.listdir(rootPath)
```

```python
imageName='id'
imageLabel='breed'
imgPath="D:\\data\\all\\trainimg\\train\\"   # location of images
filename='dogBreeds'
def dataSetGeneration(df,imageName,imageLabel,rootPath,imgPath):
   t=df.breed.unique()
   for f in t:
      os.makedirs(rootPath+filename+'\\train\\'+f)
      os.makedirs(rootPath+filename+'\\valid\\'+f)

   a=round(df.shape[0]*0.7)
   b=round(df.shape[0])

   for index, row in df[0:a].iterrows():
    print (row[imageLabel], row[imageName])
    shutil.move(imgPath+row[imageName]+'.jpg', rootPath+filename+"\\train\\"+row[imageLabel]+"\\"+row[imageName]+'.jpg')
  
   for index, row in df[a+1:b].iterrows():
    print (row[imageLabel], row[imageName])
    shutil.move(imgPath+row[imageName]+'.jpg', rootPath+filename+"valid\\"+row[imageLabel]+"\\"+row[imageName]+'.jpg')
   
  
```

```python
dataSetGeneration(data,imageName,imageLabel,rootPath,imgPath)
```
